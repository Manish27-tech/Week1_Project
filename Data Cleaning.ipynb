#**In this I am converting categorical Columns into numerical Columns where yes=1 and No=0**


```python
for col in df.select_dtypes(include='object').columns:
    print(col)
```

```python
binary_col=['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea']
for col in binary_col:
    if col in df.columns:
        df[col]=df[col].map({"yes":1,
                             "no":0})
df
```

```python
df=pd.get_dummies(df,columns=['furnishingstatus'],dtype=int)
```
